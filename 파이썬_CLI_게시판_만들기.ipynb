```python
from datetime import datetime


print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소

def find_article(num):
  cnt = 0
  for i in articles:
    if i['id'] == num:
      cnt += 1
  return cnt

def write_article(n):
    global articles
    id = n
    title = input("제목 : ")
    body = input("내용 : ")
    now = datetime.now()
    t = now.strftime("%Y-%m-%d %H:%M:%S")
    article = {"id" : id, "title" : title, "body" : body,"time" : t}
    articles.append(article)

def list_article():
  if not articles:
        print("게시글이 없습니다")
  else:
    print("=========================================")
    print("   번호      /     제목       /     내용   ")
    for a in reversed(articles):
      print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
    print("=========================================")

def delete_article(n):
  for i in articles:
    if i['id'] == n:
      articles.remove(i)

def edit_article(n):
  global articles
  for a in articles:
    if a['id'] == n:
      print("기존 제목 : {}".format(a['title']))
      print("기존 제목 : {}".format(a['body']))
      title = input("새 제목 : ")
      body = input("새 내용 : ")
      now = datetime.now()
      t = now.strftime("%Y-%m-%d %H:%M:%S")
      article = {"id" : n, "title" : title, "body" : body,"time" : t}
      articles.remove(a)
      articles.append(article)
      articles = sorted(articles,key = lambda s: s['id'])

def detail_article(n):
  for a in articles:
    if a['id'] == n:
      print("번호 : {}".format(a['id']))
      print("날짜 : {}".format(a['time']))
      print("제목 : {}".format(a['title']))
      print("내용 : {}".format(a['body']))

while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "article write":
      last_article_id += 1
      write_article(last_article_id)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다")
      else:
        list_article()


    elif cmd.startswith("article delete"):
      cmdBits = cmd.split(" ")
      if len(cmdBits) < 3:
        continue
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 글은 존재하지 않습니다.".format(index))
      else:
        delete_article(index)
        print("{}번 글이 삭제되었습니다.".format(index))

    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")
      if len(cmdBits) < 3:
        continue
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        edit_article(index)
        print("수정이 완료되었습니다.")

    elif cmd.startswith("article detail"):
      cmdBits = cmd.split(" ")
      if len(cmdBits) < 3:
        continue
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        detail_article(index)

    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")

```
