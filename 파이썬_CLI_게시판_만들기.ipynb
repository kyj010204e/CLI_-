```python
# 글 작성시간 기록을 위한 라이브러리 / 한국시간설정을 위한 라이브러리
from datetime import datetime
from pytz import timezone


print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소

# 해당 글번호의 글이 있는지 확인하는 함수: 존재 = return 1, 존재x = return 0
def find_article(num):
  cnt = 0
  for i in articles:
    if i['id'] == num:
      cnt += 1
  return cnt

# articles에 새로운 글을 작성하는 함수 n은 글번호
def write_article(n):
    global articles
    id = n
    title = input("제목 : ")
    body = input("내용 : ")
    now = datetime.now(timezone('Asia/Seoul'))
    t = now.strftime("%Y-%m-%d %H:%M:%S")
    article = {"id" : id, "title" : title, "body" : body,"time" : t}
    articles.append(article)

# artilcls에 저장된 글을 출력하는 함수
def list_article():
  if not articles:
        print("게시글이 없습니다")
  else:
    print("=========================================")
    print("   번호      /     제목       /     내용   ")
    for a in reversed(articles):
      print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
    print("=========================================")

# 글번호가 n인 글을 articles에서 찾아 삭제하는 함수
def delete_article(n):
  for i in articles:
    if i['id'] == n:
      articles.remove(i)

# 글번호가 n인 글을 articles에서 찾아 수정하는 함수
def edit_article(n):
  global articles
  for a in articles:
    if a['id'] == n:
      print("기존 제목 : {}".format(a['title']))
      print("기존 제목 : {}".format(a['body']))
      title = input("새 제목 : ")
      body = input("새 내용 : ")
      now = datetime.now(timezone('Asia/Seoul'))
      t = now.strftime("%Y-%m-%d %H:%M:%S")
      article = {"id" : n, "title" : title, "body" : body,"time" : t}
      articles.remove(a)
      articles.append(article)
      articles = sorted(articles,key = lambda s: s['id'])

# 글번호가 n인 글의 세부정보(제목, 내용, 시간, 글번호 등)를 출력하는 함수
def detail_article(n):
  for a in articles:
    if a['id'] == n:
      print("번호 : {}".format(a['id']))
      print("날짜 : {}".format(a['time']))
      print("제목 : {}".format(a['title']))
      print("내용 : {}".format(a['body']))


# 사용자의 입력을 받기위해 무한루프(exit 입력시 무한루프 종료)
while True:
    # 사용자의 입력을 저장할 cmd 변수 선언
    cmd = input("명령어 ) ").strip()
    # 사용자의 입력이 exit일 때, 무한루프 종료
    if cmd == "exit":
      break
    # 사용자의 입력이 "article write"일 때, write_article 함수 실행
    elif cmd == "article write":
      last_article_id += 1
      write_article(last_article_id)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    # 사용자의 입력이 "article list"일 때, list_aritlce 함수 실행
    elif cmd == "article list":
      # 글이 하나도 존재하지 않다면 "게시글이 없습니다" 출력
      if not articles:
        print("게시글이 없습니다")
      else:
        list_article()

    # 사용자의 입력이 "article delte"일 때, delete_article 함수 실행
    elif cmd.startswith("article delete"):
      # 사용자의 입력을 split()함수로 쪼개 마지막 요소를 index 변수로 활용
      cmdBits = cmd.split(" ")
      # 충분한 인자를 받지 못했을 경우 다시 사용자의 입력을 받음
      if len(cmdBits) < 3:
        print("명령어가 충분하지 않습니다.")
        continue
      index = int(cmdBits[2])
      # 입력받은 번호의 글이 존재하지 않는다면 삭제하지 않고 이를 사용자에게 출력
      if find_article(index) == 0:
        print("{}번 글은 존재하지 않습니다.".format(index))
      else:
        delete_article(index)
        print("{}번 글이 삭제되었습니다.".format(index))

    # 사용자의 입력이 "article edit"일 때, edit_article 함수 실행
    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")
      if len(cmdBits) < 3:
        print("명령어가 충분하지 않습니다.")
        continue
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        edit_article(index)
        print("수정이 완료되었습니다.")

    #사용자의 입력이 "article detail"일 때, detail_article 함수 실행
    elif cmd.startswith("article detail"):
      cmdBits = cmd.split(" ")
      if len(cmdBits) < 3:
        print("명령어가 충분하지 않습니다.")
        continue
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        detail_article(index)

    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")

```
