```python
print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소

def find_article(num):
  cnt = 0
  for i in articles:
    if i['id'] == num:
      cnt += 1
  return cnt

while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "article write":
      last_article_id += 1
      id = last_article_id
      title = input("제목 : ")
      body = input("내용 : ")
      article = {"id" : id, "title" : title, "body" : body}
      articles.append(article)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다")
      else:
        print("=========================================")
        print("   번호      /     제목       /     내용   ")
        for a in reversed(articles):
          print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
        print("=========================================")
    elif cmd.startswith("article delete"):
      cmdBits = cmd.split(" ")
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        for i in articles:
          if i['id'] == index:
            articles.remove(i)
        print("삭제가 완료되었습니다.")

    elif cmd.startswith("article edit"):
      cmdBits = cmd.split(" ")
      index = int(cmdBits[2])
      if find_article(index) == 0:
        print("{}번 게시글은 없습니다.".format(index))
      else:
        for a in articles:
          if a['id'] == index:
            print("기존 제목 : {}".format(a['title']))
            print("기존 제목 : {}".format(a['body']))
        title = input("새 제목 : ")
        body = input("새 내용 : ")
        article = {"id" : index, "title" : title, "body" : body}
        for i in articles:
          if i['id'] == index:
            articles.remove(i)
        articles.append(article)
        articles = sorted(articles,key = lambda s: s['id'])
        print("수정이 완료되었습니다.")
    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")

```
