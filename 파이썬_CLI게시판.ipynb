```python
# 파이썬으로 게시판 만들기
# 게시글, 회원
# 게시글 생성/수정/삭제/보기(썸네일, 상세보기)
# 회원 생성/보기
user = {

}

count = 1
print("== CLI 게시판 실행 ==")
while True:
  command = input("명령어 ) ")
  if(command == 'exit'):
    print("== CLI 게시판 종료 ==")
    break

  elif(command == 'article write'):
    title = input("제목: ")
    detail = input("내용: ")
    print("{}번째 글이 생성 되었습니다.".format(count))
    user[count] = title,detail
    count += 1


  else:
    print("지원하지 않는 기능입니다.")
    continue

print(user)
```
