```python
# 글번호, 제목, 내용을 저장할 리스트 생성
user = []

# 리스트에서 원하는 글번호의 딕셔너리를 삭제하는 함수
def delete_user(n):
  for i in user:
          if i['id']== n:
            user.remove(i)

# 리스트에서 원하는 글번호가 존재하는지 확인하는 함수 존재하면 return = 1 존재하지 않으면 return 0
def find_user(n):
  cnt = 0
  for i in user:
    if i['id'] == n:
      cnt += 1
  return


# 입력받은 명령어가 인자를 필요로하는지 판단하는 함수
def command_split(cmd):
  if len(command.split(' ')) >= 3:
    return 3
  else:
    return 2


# 행기준 user[마지막] 부터 user[0] 까지 각 요소를 출력하는 함수
def article_list(cnt):
  print("==============================================================================")
  print("번호","제목","내용",sep = '\t\t/\t\t')
  for i in reversed(user):
    print(i['id'],end = '\t\t/\t\t')
    print(i['title'],end = '\t\t/\t\t')
    print(i['detail'],end ='\n')
  print("==============================================================================")
def article_list_one(cnt):
  print("==============================================================================")
  print("번호","제목","내용",sep = '\t\t/\t\t')
  for i in user:
    if i['id'] == cnt:
      print(i['id'],end = '\t\t/\t\t')
      print(i['title'],end = '\t\t/\t\t')
      print(i['detail'],end ='\n')
  print("==============================================================================")


# 글번호를 구분하기 위한 변수 count
count = 0



print("== CLI 게시판 실행 ==")
while True:
  # 초기에는 항상 command를 입력받음, strip함수를 통해 처음과 마지막의 공백은 무시
  command = input("명령어 ) ").strip()
  command_len = command_split(command)

  if command_len == 2:
  # command가 exit일때, 프로그램 종료
    if command == 'exit':
      print("== CLI 게시판 종료 ==")
      break

    # commnad가 article write일때 글작성 이후 글번호를 1 늘림
    elif command == 'article write':
      count += 1
      title = input("제목: ")
      detail = input("내용: ")
      article = {'id' : count, 'title' : title, 'detail' : detail}
      user.append(article)
      print("{}번째 글이 생성 되었습니다.".format(count))

    # command가 article list일때 article list 함수에 count 인자를 넣고 실행
    elif(command == 'article list'):
      if len(user) == 0:
        print("작성된 글이 없습니다.")
      else:
        article_list(count)
    else:
      print("지원하지 않는 기능입니다.")
      continue

  # command가 인자를 더 필요로할 때
  elif command_len == 3:
    cmd = command.split(' ')

    # article delete를 입력했을 때 해당 글번호를 삭제
    if cmd[0]+cmd[1] == 'articledelete':
      n= int(cmd[2])

      if find_user(n)== 0:
        print("{}번글은 존재하지 않습니다.".format(n))
      else:
        delete_user(n)
        print("{}번 글이 삭제되었습니다.".format(n))

    # article edit을 입력했을 때 해당 글번호를 수정
    elif cmd[0]+cmd[1] == 'articleedit':
      n= int(cmd[2])
      if find_user(n)== 0:
        print("{}번글은 존재하지 않습니다.".format(n))
      else:
        title = input("제목: ")
        detail = input("내용: ")
        article = {'id' : n, 'title' : title, 'detail' : detail}
        delete_user(n)
        user.append(article)
        print("수정되었습니다.")
        # user = sorted(user,key = lambda s: s['id'])

    # article detail을 입력했을 때 해당 글번호를 상세히 보여줌
    elif cmd[0]+cmd[1] == 'articledetail':
      n= int(cmd[2])
      if find_user(n)== 0:
        print("{}번글은 존재하지 않습니다.".format(n))
      else:
        article_list_one(n)

    else:
      print("지원하지 않는 기능입니다.")
      continue


```
