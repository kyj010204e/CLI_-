```python
# 파이썬으로 게시판 만들기
# 게시글, 회원
# 게시글 생성/수정/삭제/보기(썸네일, 상세보기)
# 회원 생성/보기
import numpy as np

# n,3의 2차원 배열 선언
user = np.array([[0,0,0]])


# 행기준 user[마지막] 부터 user[0] 까지 각 요소를 출력
def article_list(cnt):
  j = 0
  print("=======================================")
  print("번호","제목","내용",sep = '\t\t/\t\t')
  for i in range(cnt-1,0,-1):
    print(user[(i,0)],end = '\t\t/\t\t')
    print(user[(i,1)],end = '\t\t/\t\t')
    print(user[(i,2)],end = '\n')

  print("=======================================")


# 글번호를 구분하기 위한 변수 count
count = 1



print("== CLI 게시판 실행 ==")
while True:
  # 초기에는 항상 command를 입력받음
  command = input("명령어 ) ")

  # command가 exit일때, 프로그램 종료
  if command == 'exit':
    print("== CLI 게시판 종료 ==")
    break

  # commnad가 article write일때 글작성 이후 글번호를 1 늘림
  elif command == 'article write' or command == '1':
    title = input("제목: ")
    detail = input("내용: ")
    print("{}번째 글이 생성 되었습니다.".format(count))

    # 생성해둔 2차열배열 user에 새로운 배열을 추가함 행기준 append
    user = np.append(user,[[count,title,detail]],axis = 0)
    count += 1

  elif(command == 'article list'):
     article_list(count)

  else:
    print("지원하지 않는 기능입니다.")
    continue
```
