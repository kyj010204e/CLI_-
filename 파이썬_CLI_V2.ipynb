```python
# n,3의 2차원 배열 선언
user = []


# 행기준 user[마지막] 부터 user[0] 까지 각 요소를 출력하는 함수
def article_list(cnt):
  print("==============================================================================")
  print("번호","제목","내용",sep = '\t\t/\t\t')
  for i in reversed(user):
    print(i['id'],end = '\t\t/\t\t')
    print(i['title'],end = '\t\t/\t\t')
    print(i['detail'],end ='\n')
  print("==============================================================================")


# 글번호를 구분하기 위한 변수 count
count = 0



print("== CLI 게시판 실행 ==")
while True:
  # 초기에는 항상 command를 입력받음, strip함수를 통해 처음과 마지막의 공백은 무시
  command = input("명령어 ) ").strip()

  # command가 exit일때, 프로그램 종료
  if command == 'exit':
    print("== CLI 게시판 종료 ==")
    break

  # commnad가 article write일때 글작성 이후 글번호를 1 늘림
  elif command == 'article write':
    count += 1
    title = input("제목: ")
    detail = input("내용: ")
    article = {'id' : count, 'title' : title, 'detail' : detail}
    user.append(article)
    print("{}번째 글이 생성 되었습니다.".format(count))



  # command가 article list일때 article list 함수에 count 인자를 넣고 실행
  elif(command == 'article list'):
     article_list(count)

  # 예상하지 못한 명령어를 입력받았을 때 해당 command를 무시
  else:
    print("지원하지 않는 기능입니다.")
    continue
```
